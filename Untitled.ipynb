```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate sample data
data = np.random.randn(10, 10)
df = pd.DataFrame(data, columns=['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'])

# Set the values to be highlighted
highlight_values = [-0.5, 1, 0.5]

# Create diverging colormap
cmap = plt.get_cmap('coolwarm', lut=256)

# Set the minimum and maximum values for the colormap
vmin, vmax = -1, 1

# Create the figure and the subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot the heatmap for correlation
corr = df.corr()
corr_im = axs[0].imshow(corr, cmap=cmap, vmin=vmin, vmax=vmax)

# Add colorbar for correlation
corr_cbar = fig.colorbar(corr_im, ax=axs[0], fraction=0.046, pad=0.04)
corr_cbar.ax.tick_params(labelsize=8)

# Plot the heatmap for bias
bias = df.corr(method='spearman')
bias_im = axs[1].imshow(bias, cmap=cmap, vmin=vmin, vmax=vmax)

# Add colorbar for bias
bias_cbar = fig.colorbar(bias_im, ax=axs[1], fraction=0.046, pad=0.04)
bias_cbar.ax.tick_params(labelsize=8)

# Add titles to the subplots
axs[0].set_title('Correlation', fontsize=10)
axs[1].set_title('Bias', fontsize=10)

# Add annotations to the subplots
for i in range(corr.shape[0]):
    for j in range(corr.shape[1]):
        corr_val = corr.iloc[i, j]
        bias_val = bias.iloc[i, j]
        if corr_val in highlight_values and bias_val in highlight_values:
            axs[0].text(j, i, '{:.2f}'.format(corr_val), ha='center', va='center', color='blue', fontsize=8, fontweight='bold')
            axs[1].text(j, i, '{:.2f}'.format(bias_val), ha='center', va='center', color='blue', fontsize=8, fontweight='bold')
        else:
            axs[0].text(j, i, '{:.2f}'.format(corr_val), ha='center', va='center', color='black', fontsize=8)
            axs[1].text(j, i, '{:.2f}'.format(bias_val), ha='center', va='center', color='black', fontsize=8)

# Show the plot
plt.suptitle('Correlation and Bias Heatmaps', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create example data
data = np.random.rand(10, 10) * 100

# Define range of values to color-code
vmin, vmax = 40, 80

# Create heatmap with colorbar
fig, ax = plt.subplots()
im = ax.imshow(data, cmap='coolwarm', vmin=vmin, vmax=vmax)

# Create colorbar
cbar = ax.figure.colorbar(im, ax=ax, fraction=0.046, pad=0.04)

# Set colorbar label and tick labels
cbar.ax.set_ylabel('Value', rotation=-90, va="bottom")
cbar.set_ticks([vmin, vmax])
cbar.set_ticklabels([str(vmin), str(vmax)])

# Add title
ax.set_title('Heatmap with Color-Coded Range')

# Show plot
plt.show()
```
